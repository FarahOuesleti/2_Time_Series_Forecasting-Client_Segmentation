```python
def DataDiscretisation(do):
  Converted_Date = []
  for l in range(len(do)) :
    #Discretisation days of the month
    if (do.loc[l]['date'].day in range(1,3))  :
      day='01'
    elif (do.loc[l]['date'].day in range(3,5)) :
      day='03'
    elif (do.loc[l]['date'].day in range(5,7)) :
      day='05'
    elif (do.loc[l]['date'].day in range(7,9)) :
      day='07'
    elif (do.loc[l]['date'].day in range(9,11)) :
      day='09'
    elif (do.loc[l]['date'].day in range(11,13)) :
      day='11'
    elif (do.loc[l]['date'].day in range(13,15)) :
      day='13'
    elif (do.loc[l]['date'].day in range(15,17)) :
      day='15'
    elif (do.loc[l]['date'].day in range(17,19)) :
      day='17'
    elif (do.loc[l]['date'].day in range(19,21)) :
      day='19'
    elif (do.loc[l]['date'].day in range(21,23)) :
      day='21'
    elif (do.loc[l]['date'].day in range(23,25)) :
      day='23'
    elif (do.loc[l]['date'].day in range(25,27)) :
      day='25'
    elif (do.loc[l]['date'].day in range(27,29)) :
      day='27'
    elif (do.loc[l]['date'].day in range(29,31)) :
      day='29'
    elif (do.loc[l]['date'].day == 31) :
      day='31'

    # Discretisation hours of the day
    for i in range(10):
      if do.loc[l]['date'].hour == i : hour = '0'+str(i)
    for i in range(10,24):
      if do.loc[l]['date'].hour == i : hour = str(i)

    m = str(do.loc[l]["date"].month)
    if len(m) ==1 :
      m='0'+m
    New_Date = day +'/'+ m +'/2023 '+ hour +':00'

    Converted_Date.append(pd.to_datetime(New_Date,format='%d/%m/%Y %H:%M'))

  do['Converted_Date'] = Converted_Date
  return 'success'
  #return Converted_Date

DataDiscretisation(do)
```

```python
def NumberByProduct(do) :
  #Sup_Fr,C_Esc,C_Thon,PM_Nep,PM_Mar,PM_R=[],[],[],[],[],[]
  Sup_caramel,Sup_nutella,Jus_Orange,Citronnade = [0 for i in range(70)],[0 for i in range(70)],[0 for i in range(70)],[0 for i in range(70)]
  Express,Direct,Turc = [0 for i in range(70)],[0 for i in range(70)],[0 for i in range(70)]
  NB_Total_Commandes = []
  NB_Delivery = []
  Total_Quantity = []
  Sum_Revenues = []
  Converted_Date = []

  if do['article.product_name'][0] == 'Supp caramel' : Sup_caramel[0] = 1
  elif do['article.product_name'][0] == 'Supp nutella' : Sup_nutella[0] = 1
  elif do['article.product_name'][0] == 'Jus d\'Orange' : Jus_Orange[0] = 1
  elif do['article.product_name'][0] == 'Citronnade' : Citronnade[0] = 1
  elif do['article.product_name'][0] == 'Express' : Express[0] = 1
  elif do['article.product_name'][0] == 'Direct' : Direct[0] = 1
  elif do['article.product_name'][0] == 'Cafe Turc' : Turc[0] = 1

  l=1
  while l < len(do) :
    Converted_Date.append(do['Converted_Date'][l])
    iCom = 1
    deliv = do['delivery'][l]
    quantity = do['article.quantity'][l]
    Revenues = do['total'][l]

    #print('pass 1 : ',l)

    while (do['Converted_Date'][l] == do['Converted_Date'][l-1]) :
      #print('pass 2 : ',iCom)
      iCom +=1
      deliv += do['delivery'][l]
      quantity += do['article.quantity'][l]
      Revenues += do['total'][l]


      if do['article.product_name'][l] == 'Supp caramel' :
        Sup_caramel[len(NB_Delivery)] += 1
      elif do['article.product_name'][l] == 'Supp nutella' :
        Sup_nutella[len(NB_Delivery)] += 1
      elif do['article.product_name'][l] == 'Jus d\'Orange' :
        Jus_Orange[len(NB_Delivery)] += 1
      elif do['article.product_name'][l] == 'Citronnade' :
        Citronnade[len(NB_Delivery)] += 1
      elif do['article.product_name'][l] == 'Express' :
        Express[len(NB_Delivery)] += 1
      elif do['article.product_name'][l] == 'Direct' :
        Direct[len(NB_Delivery)] += 1
      elif do['article.product_name'][l] == 'Cafe Turc' :
        Turc[len(NB_Delivery)] += 1

      l+=1
      if (l==321) : break
    l+=1

    #Save summary data for new line in group G(x,y) where x: day phase in the month [1..6], y hour phase in the day [1..6]
    NB_Total_Commandes.append(iCom)
    NB_Delivery.append(deliv)
    Total_Quantity.append(quantity)
    Sum_Revenues.append(Revenues)

  dictClean = {'Converted_Date' : Converted_Date, 'Sum_Revenues' : Sum_Revenues,'Total_Quantity' : Total_Quantity, 'NB_Delivery' : NB_Delivery,'NB_Total_Commandes' :NB_Total_Commandes ,
          'Sup_caramel':Sup_caramel,'Sup_nutella':Sup_nutella,'Jus_Orange':Jus_Orange,'Citronnade':Citronnade,'Express':Express,'Direct':Direct,'Turc':Turc}


  # Make all lists same length as that of Converted_Dates to make a DF out of them eventually
  for k,v in dictClean.items() :
    if k not in ['Converted_Date','Total_Quantity','NB_Delivery','NB_Total_Commandes','Sum_Revenues']:
      dictClean[k]=v[:len(Converted_Date)]


  dc = pd.DataFrame(data=dictClean, index=Converted_Date)

  return dc

Clean_Discrete_Drinks = NumberByProduct(do)
```
